# определение типов данных

```python
import pandas as pd
import numpy as np

# Укажите путь к вашему файлу
file_path = r'C:\laba-python\victory\filtered_data_pq\catalog.pq'

# Чтение файла Parquet
df = pd.read_parquet(file_path)

# Выводим точные типы данных в каждом столбце
for column in df.columns:
    # Проверяем если dtype это object
    if df[column].dtype == 'object':
        # Применяем type к каждому элементу в столбце
        unique_types = df[column].apply(lambda x: type(x) if isinstance(x, (list, np.ndarray)) else type(x)).unique()
        print(f"Column: {column}, Unique types: {unique_types}")
    else:
        print(f"Column: {column}, Numpy type: {df[column].dtype}")
```
